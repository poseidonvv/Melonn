#**Queremos ver los diferentes tipos e pedidos por año mes o por día **


```python
dataClassMes = pd.crosstab(df.RequestTime, df.DeliveryOption).resample('M').sum() #separamos por horario de entrega a mes
dataClassDay = pd.crosstab(df.RequestTime, df.DeliveryOption).resample('D').sum() #separamos por horario de entrega a dia 
dataClassDay['num_envi_mornin'] = [0 if i-1 == 0 else dataClassDay['Siguiente dia entre 12:30 pm y 2:30 pm'][i-2] + dataClassDay['Siguiente dia entre las 6:30 pm y 8:30 pm'][i-2] for i in range(1, len(dataClassDay)+1) ]
dataClassDay['num_envi_afther'] = [dataClassDay['Mismo dia entre 6:30 pm y 8:30 pm'][i] for i in range(len(dataClassDay))]
dataClassDay #pedidos por día, jornada mañana y tarde
```

```python
dataClassDay['date'] = dataClassDay.index.values
dataClassDay['dateMonth'] = pd.DatetimeIndex(dataClassDay['date']).month
dataClassDay['dateYear'] = pd.DatetimeIndex(dataClassDay['date']).year
dtyear = dataClassDay.groupby(['dateYear', 'dateMonth']) # defino grupos de la forma (año, #mes)
a= []  ## medias de num_env_mornin
b= []  ## medias de num_env_afther
maxTurnMornin = [] # máximo por mes turno mañana 
maxTurnAfther = [] # máximo por mes turno tarde
for names, group in dtyear:
  a.append(dtyear.get_group (names)['num_envi_mornin'].max())
  b.append(dtyear.get_group (names)['num_envi_afther'].max())
  maxTurnMornin.append(dtyear.get_group (names)['num_envi_mornin'].max())
  maxTurnAfther.append(dtyear.get_group (names)['num_envi_afther'].max())

dataClassMes['MaxEnvMornin'] = a
dataClassMes['MaxEnvAfthern'] = b
dataClassMes['maxPeopleTurnMornin'] = [(maxTurnMornin[i]//5)+1 for i in range(len(dataClassMes))  ] # mínimo de personal necesitado turno mañana
dataClassMes['maxpeopleTurnAfther'] = [(maxTurnAfther[i]//5)+1 for i in range(len(dataClassMes))  ] # mínimo de personal necesitado turno tarde
dataClassMes['impact'] = impact
dataClassMes['date'] = dataClassMes.index.values
dataClassMes = pd.DataFrame(data = dataClassMes)
dataClassMes

```
